```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Convert date column
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Create Month and Quarter columns
df["Month"] = df["Order Date"].dt.to_period("M")
df["Quarter"] = df["Order Date"].dt.to_period("Q")
```

```python
monthly_revenue = df.groupby("Month")["Revenue"].sum()

plt.figure()
monthly_revenue.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("monthly_revenue_trend.png")
plt.show()
```

```python
quarterly_revenue = df.groupby("Quarter")["Revenue"].sum()

plt.figure()
quarterly_revenue.plot()
plt.title("Quarterly Revenue Trend")
plt.xlabel("Quarter")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("quarterly_revenue_trend.png")
plt.show()
```

```python
category_revenue = df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)

plt.figure()
category_revenue.plot(kind="bar")
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("category_revenue_bar.png")
plt.show()
```

```python
plt.figure()
category_revenue.plot(kind="pie", autopct="%1.1f%%")
plt.title("Revenue Share by Category")
plt.ylabel("")
plt.tight_layout()
plt.savefig("category_revenue_pie.png")
plt.show()
```
